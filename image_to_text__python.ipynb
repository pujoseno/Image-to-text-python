```
import cv2
from google.colab.patches import cv2_imshow
```

```
originalImage = cv2.imread('anjay.png')
cv2_imshow(originalImage)
```

```
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImage)
```

```
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 240, 255, cv2.THRESH_BINARY)
cv2_imshow(blackAndWhiteImage)
print(thresh)
```

```
convertimg= pytesseract.image_to_string(blackAndWhiteImage)
```
